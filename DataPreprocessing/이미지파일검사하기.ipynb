```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# 디렉토리 리스트
directories = [
    "D:\\data\\tomato\\SourceData\\71_data\\Training\\0_healthy",
    "D:\\data\\tomato\\SourceData\\71_data\\Training\\1_disease",
    "D:\\data\\tomato\\SourceData\\71_data\\Validation\\0_healthy",
    "D:\\data\\tomato\\SourceData\\71_data\\Validation\\1_disease",
    
    "D:\\data\\tomato\\SourceData\\73_data\\Training\\0_healthy",
    "D:\\data\\tomato\\SourceData\\73_data\\Validation\\0_healthy",
    
    "D:\\data\\tomato\\SourceData\\104_data\\Training\\0_healthy",
    "D:\\data\\tomato\\SourceData\\104_data\\Training\\1_disease",
    "D:\\data\\tomato\\SourceData\\104_data\\Validation\\0_healthy",
    "D:\\data\\tomato\\SourceData\\104_data\\Validation\\1_disease",
    
    "D:\\data\\tomato\\SourceData\\PlantVillage\\0_healthy",
    "D:\\data\\tomato\\SourceData\\PlantVillage\\1_disease\\3_YellowLeafCurlVirus",
    "D:\\data\\tomato\\SourceData\\PlantVillage\\1_disease\\4_LeafMold",
    "D:\\data\\tomato\\SourceData\\PlantVillage\\1_disease\\5_EarlyBlight",
    
    "D:\\data\\tomato\\SourceData\\SmartFarm\\1_disease\\1_PowderyMildew",
    "D:\\data\\tomato\\SourceData\\SmartFarm\\1_disease\\3_YellowLeafCurlVirus",
    "D:\\data\\tomato\\SourceData\\SmartFarm\\1_disease\\4_LeafMold"
]


error_files = []  # 오류가 발생한 이미지 파일들의 경로를 저장할 리스트

# 이미지의 새로운 크기를 설정
new_width = 256
new_height = 256

# 모든 디렉토리를 순회
for directory in directories:
    for root, _, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                image_path = os.path.join(root, file)
                try:
                    with Image.open(image_path).resize((new_width, new_height)) as img:
                        plt.imshow(img)
                        pass  # 이미지 오류 검사만 수행하고 출력은 하지 않음
                except Exception as e:
                    print(f"오류가 발생한 이미지 경로: {image_path}")
                    error_files.append(image_path)

# 오류가 발생한 이미지 파일들의 경로를 .txt 파일에 저장
with open("G:/error_image_paths.txt", "w") as file:
    for path in error_files:
            file.write(path + "\n")

print(f"오류가 발생한 이미지 파일 {len(error_files)}개의 경로가 'error_image_paths.txt'에 저장되었습니다.")
```
