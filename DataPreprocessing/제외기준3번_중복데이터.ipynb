```python
import os

# 경로 설정
path = r"D:\data\tomato\SourceData\71_data\Training\0_healthy"

# 파일 목록 가져오기
file_list = os.listdir(path)
filtered_files = []

# 13번째 언더스코어 이후에 추가 요소가 있는 파일 찾기
for file in file_list:
    parts = file.split('_')
    if len(parts) == 12:  # 언더스코어가 12개일때
        filtered_files.append(file)  # 리스트에 추가
        print(file)  # 해당 파일명 출력
```

```python
# 리스트 예제

# 파일에 저장하기
with open("D:/list_output.txt", "w", encoding="utf-8") as file:
    for item in filtered_files:
        file.write("%s\n" % item)
```

```python
print(len(filtered_files))
```
