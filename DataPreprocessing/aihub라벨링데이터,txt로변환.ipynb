```python
#71_data\Training#건강
import os
import json

# 경로 정의
json_path = r"D:\data\tomato\SourceData\labeling_AihubTotal\71_data\Training\0_healthy"
image_path = r"D:\data\tomato\SourceData\71_data\Training\0_healthy"
output_path = r"D:\data\tomato\SourceData\71_data\Training\labeling"

invalid_data_files = []  # YOLO 형식으로 변환할 수 없는 데이터가 있는 JSON 파일 리스트

# 이미지 파일들에 대해서만 처리
for image_file in os.listdir(image_path):
    if image_file.endswith(('.jpg', '.png', '.jpeg', '.JPG')):
        corresponding_json = image_file.replace('.jpg', '.jpg.json').replace('.JPG', '.jpg.json')

        # 해당 JSON 파일이 존재하는 경우만 처리
        if os.path.exists(os.path.join(json_path, corresponding_json)):
            with open(os.path.join(json_path, corresponding_json), 'r') as f:
                data = json.load(f)

            try:
                # 이미지의 너비와 높이
                width = data["description"]["width"]
                height = data["description"]["height"]

                # 좌표 추출
                xtl = data["annotations"]["points"][0]["xtl"]
                ytl = data["annotations"]["points"][0]["ytl"]
                xbr = data["annotations"]["points"][0]["xbr"]
                ybr = data["annotations"]["points"][0]["ybr"]

                # YOLO 형식 변환
                x_center = (xtl + xbr) / 2.0
                y_center = (ytl + ybr) / 2.0
                w = xbr - xtl
                h = ybr - ytl

                # 좌표 정규화
                x_center = round(x_center / width, 6)
                y_center = round(y_center / height, 6)
                w = round(w / width, 6)
                h = round(h / height, 6)

                # 출력 파일 이름 설정
                txt_file_name = corresponding_json.replace('.json', '.txt')
                with open(os.path.join(output_path, txt_file_name), 'w') as out_file:
                    out_file.write(f"0 {x_center} {y_center} {w} {h}\n")
            except:
                invalid_data_files.append(corresponding_json)
        else:
            missing_json_files.append(image_file)
```
