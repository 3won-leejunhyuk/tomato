```python
import os
import glob
from PIL import Image, ExifTags

orig_dir = r"F:\data\tomato\그린데이터랩\tomato_except4\TestDataSet\PlantVillage\5_EarlyBlight"

photos = []

# Loop through all the supported extensions and add the file paths to the photos list.
for ext in os.listdir(orig_dir):
    if ext.lower().endswith(('jpg', 'png', 'jpeg')):
        photos.append(os.path.join(orig_dir, ext))
print("total number of photos is: " + str(len(photos)))
```

```python
for photo in photos:
    try:
        print(photo)
        if not os.path.isfile(photo):
            continue
        
        im = Image.open(photo)

        for orientation in ExifTags.TAGS.keys():
            if ExifTags.TAGS[orientation] == 'Orientation':
                break

        exif = dict(im._getexif().items())

        if exif[orientation] == 3:
            im = im.rotate(180, expand=True)
        elif exif[orientation] == 6:
            im = im.rotate(270, expand=True)
        elif exif[orientation] == 8:
            im = im.rotate(90, expand=True)

        im.save(photo) # Save the image with the same name (overwriting the original one).
        
    except Exception as e:
        print("caught error: ", e)

    finally:
         im.close()
```

```python
for photo in photos:
    try:
        print(photo)
        if not os.path.isfile(photo):
            continue

        im = Image.open(photo)
        
        # Get the file extension
        _, file_extension = os.path.splitext(photo)

        # If the file is not a jpg, save it as a jpg
        if file_extension.lower() != '.jpg':
            new_file = photo.replace(file_extension, '.jpg')
            im.save(new_file, 'JPEG')
            os.remove(photo)
        else:
            im.save(photo)  # Save the image with the same name (overwriting the original one).
            
    except Exception as e:
        print("caught error: ", e)
        
    finally:
        im.close()
```
