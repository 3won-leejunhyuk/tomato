```python
import os
import csv

# Define the base directory paths
base_directories = {
    'TrainingDataSet': '/workspace/mnt2/data/tomato_except3/leaf/TrainingDataSet/',
    'ValidationDataSet': '/workspace/mnt2/data/tomato_except3/leaf/ValidationDataSet/',
    'TestDataSet': '/workspace/mnt2/data/tomato_except3/leaf/TestDataSet'    
}
csv_file_path = '/workspace/mnt2/data/tomato_except3/total_label_counts.csv'

# Define the subdirectories
subdirectories = ['0_healthy', '1_PowderyMildew', '2_GrayMold', '3_YellowLeafCurlVirus', '4_LeafMold']

# Function to count labels
def count_labels(directory):
    label_counts = {}
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            filepath = os.path.join(directory, filename)
            with open(filepath, 'r') as file:
                for line in file:
                    label = line.strip().split()[0]
                    label_counts[label] = label_counts.get(label, 0) + 1
    return label_counts

# Initialize a dictionary to store total counts
total_counts = {'Total labels for all datasets combined': 0}

# Compute total counts for each dataset and subdirectory
for dataset_name, dataset_path in base_directories.items():
    total_counts[f'Total labels for {dataset_name}'] = 0  # Initialize dataset total count
    for subdirectory in subdirectories:
        subdirectory_path = os.path.join(dataset_path, subdirectory)
        subdirectory_label_counts = count_labels(subdirectory_path)
        subdirectory_total = sum(subdirectory_label_counts.values())
        total_counts[f'{subdirectory} label'] = subdirectory_total  # Store subdirectory count
        total_counts[f'Total labels for {dataset_name}'] += subdirectory_total  # Update dataset total count
        total_counts['Total labels for all datasets combined'] += subdirectory_total  # Update grand total count

# Write the totals to a CSV file
with open(csv_file_path, mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in total_counts.items():
        writer.writerow([key, value])

# Print the path to the output CSV file
print(f"Total label counts have been written to {csv_file_path}")
```

```python
import os

# Define the base directory for the training, test, and validation datasets
base_directories = {
    'TrainingDataSet': '/workspace/mnt2/data/tomato_except3/leaf/5/TrainingDataSet/',
    'ValidationDataSet': '/workspace/mnt2/data/tomato_except3/leaf/5/ValidationDataSet/',
    'TestDataSet': '/workspace/mnt2/data/tomato_except3/leaf/5/TestDataSet'    
}

# Define the subdirectories as shown in your folder structure
subdirectories = ['0_healthy', '1_PowderyMildew', '2_GrayMold', '3_YellowLeafCurlVirus', '4_LeafMold', '5_LateBlight']

# This function will count the number of labels in YOLO formatted .txt files within a directory
def count_labels(directory):
    label_counts = {}
    # Only proceed if the directory exists
    if os.path.exists(directory):
        for filename in os.listdir(directory):
            if filename.endswith('.txt'):
                filepath = os.path.join(directory, filename)
                with open(filepath, 'r') as file:
                    for line in file:
                        label = line.strip().split()[0]
                        label_counts[label] = label_counts.get(label, 0) + 1
    return label_counts

# Dictionary to store the counts for each subdirectory in each main dataset directory
dataset_label_counts = {}
total_labels_all_datasets = 0

# Iterate through each dataset directory and their respective subdirectories
for dataset, base_dir in base_directories.items():
    dataset_label_counts[dataset] = {}
    dataset_total_labels = 0
    for subdirectory in subdirectories:
        full_path = os.path.join(base_dir, subdirectory)
        subdirectory_label_counts = count_labels(full_path)
        dataset_label_counts[dataset][subdirectory] = subdirectory_label_counts
        # Sum the labels for the current subdirectory and add to the dataset total
        dataset_total_labels += sum(subdirectory_label_counts.values())
    # Add the dataset total to the grand total
    total_labels_all_datasets += dataset_total_labels
    # Print the total for the current dataset
    print(f"Total labels for {dataset}: {dataset_total_labels}")

# Print the total counts for each dataset
for dataset, subdirs in dataset_label_counts.items():
    print(f"Label counts for {dataset}:")
    for subdir, counts in subdirs.items():
        print(f" - {subdir}: {sum(counts.values())} labels")

# Print the combined total for all datasets
print(f"Total labels for all datasets combined: {total_labels_all_datasets}")
```

```python
import os

def count_labels_in_yolo_format(root_path):
    # A dictionary to hold the count of each label
    label_counts = {}

    # Walk through the directory
    for subdir, dirs, files in os.walk(root_path):
        for file in files:
            if file.endswith('.txt'):  # Check if the file is a .txt file
                file_path = os.path.join(subdir, file)
                with open(file_path, 'r') as f:
                    for line in f:
                        if line.strip():  # Check if the line is not empty
                            label = line.split()[0]  # The first element is the label
                            label_counts[label] = label_counts.get(label, 0) + 1

    return label_counts

# Assuming the root path is where the leaf folder is located
root_path = '/workspace/mnt2/data/tomato_except3/leaf/TestDataSet/'
labels_count = count_labels_in_yolo_format(root_path)
print(labels_count)
```
