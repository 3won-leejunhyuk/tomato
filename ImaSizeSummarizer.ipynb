# 이미지

```python
import os
import csv
from PIL import Image
```

```python
# Define base directories and subdirectories
base_directories = {
    'TrainingDataSet': '/workspace/mnt2/data/tomato_except3/crop/0/crop/TrainingDataSet/',
    'ValidationDataSet': '/workspace/mnt2/data/tomato_except3/crop/0/crop/ValidationDataSet/',
}
subdirectories = ['0_healthy', '1_PowderyMildew', '2_GrayMold', '3_YellowLeafCurlVirus', '4_LeafMold', '5_LateBlight']

# Path for the CSV file
csv_file_path = '/workspace/mnt2/data/tomato_except3/crop/0/crop/total_img_size.csv'

# Prepare to collect image data
image_data = []

# Iterate over the base directories and subdirectories
for base_dir_name, base_dir_path in base_directories.items():
    for sub_dir in subdirectories:
        sub_dir_path = os.path.join(base_dir_path, sub_dir)
        # Check if subdirectory exists to avoid errors
        if os.path.exists(sub_dir_path):
            for image_file in os.listdir(sub_dir_path):
                image_path = os.path.join(sub_dir_path, image_file)
                # Open the image to get its size
                with Image.open(image_path) as img:
                    width, height = img.size
                    # Append the data to the list
                    image_data.append([base_dir_name, sub_dir, image_file, width, height])

# Write the collected data to a CSV file
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write the header
    writer.writerow(['BaseDirectory', 'SubDirectory', 'FileName', 'Width', 'Height'])
    # Write the image data
    writer.writerows(image_data)
```

# 라벨

```python
import os
from PIL import Image
import csv

# Define the base directory containing the images and labels
base_directory = '/workspace/mnt/data/tomato_except4_revise/TestDataSet/'

# CSV 파일 이름 정의
csv_file_name = '/workspace/mnt/data/tomato_except4_revise/TestDataSetLabel_sizes.csv'

# CSV 파일 헤더
headers = ['Image Name', 'Class ID', 'Label Width', 'Label Height']
valid_extensions = ('.jpg', '.jpeg', '.png')
# CSV 파일 생성 및 헤더 작성
with open(csv_file_name, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(headers)

    # 이미지와 라벨 파일 탐색
    for root, dirs, files in os.walk(base_directory):
        for file in files:
            if file.lower().endswith(valid_extensions):
                image_path = os.path.join(root, file)
                print(image_path)
                label_path = image_path.rsplit('.', 1)[0] + '.txt'
                
                if os.path.exists(label_path):
                    with Image.open(image_path) as img:
                        img_width, img_height = img.size
                        
                        with open(label_path, 'r') as label_file:
                            for line in label_file:
                                class_id, x_center, y_center, width, height = map(float, line.split())
                                
                                # YOLO 형식의 상대 크기를 실제 픽셀 크기로 변환
                                absolute_width = width * img_width
                                absolute_height = height * img_height
                                
                                # CSV 파일에 라인 작성
                                writer.writerow([file, int(class_id), absolute_width, absolute_height])
```
